#WHAT IS IT?:
#A heatmap is a way to visualize a table of numbers
#A heatmap is a way to vizualize data that presents magnitude of features as color in two dimensions. 
#The variation in color in a heatmap may be by hue or intensity,
which gives obvious visual cues to a reader about how the feature is clustered or varies over space

#A heatmap is a representation of data in the form of a map or diagram in which data values are represented as colors

#An image or map representing the varying features

#A heatmap depicts values for a main variable of interest across two axis variables as a grid of colored squares.
The axis variables are divided into ranges like a bar chart or histogram, 
#and each cellâ€™s color indicates the value of the main variable in the corresponding cell range

#WHAT IS IT USED FOR?:
#To show how features are clustered or the associations/ relationships between two variables/features
#Used for visualizing a correlation matrix

#WHEN IS IT USED?: 
#A heat map helps you visualize density. So use them when you want to show how data is clustered
#Heatmaps are used to show relationships between two variables, one plotted on each axis. 
#By observing how cell colors change across each axis, you can observe if there are any patterns in value for one or both variables

```python
#For Week 6, Day 3 we will work with two kinds of dataset to create a heatmap (1) a regular dataframe, and (2) a numpy array
```

```python
#1. Using a regular dataframe

import seaborn as sns
import pandas as pd
educ_data = pd.read_csv("https://vincentarelbundock.github.io/Rdatasets/csv/AER/CPSSW8.csv", index_col=0).dropna()
educ_data.head()
```

```python
educ_data.describe().T
```

```python
# check correlations for educ_data dataset

educ_data.corr() # this gives you the Pearson correlation coefficient
```

```python
#Plot a heatmap for the educ_data dataframe

sns.heatmap(educ_data.corr())
```

```python
#Set the limits of the correlation matrix heatmap between -1 and 1

sns.heatmap(educ_data.corr(), center=0, vmin=-1, vmax=1)
```

```python
#2. Using a numpy array

#Plot a heatmap for a numpy array

import numpy as np; np.random.seed(2)
import seaborn as sns; sns.set_theme()
random_data = np.random.rand(11, 13)
sns.heatmap(random_data)
```

```python
sns.heatmap(random_data, center=0, vmin=-1, vmax=1)
```

```python
#Loading Matplotlib and Seaborn

from matplotlib import pyplot as plt
import seaborn as sns
```

```python
cars = sns.load_dataset('mpg')
```

```python
cars.head()
```

```python
cars.groupby('origin').cylinders.value_counts()
```

```python
#Unstack your data

cars.groupby('origin').cylinders.value_counts().unstack()

```

```python
#Fill NaNs with zeroes
#Some version of Seaborn accepts NaNs but you end up with blank heatmap cells

cars.groupby('origin').cylinders.value_counts().unstack().fillna(0)
```

```python
#Assign table to an object name


Origin_Cylinder = cars.groupby('origin').cylinders.value_counts().unstack().fillna(0)
```

```python
sns.heatmap(Origin_Cylinder)
```

```python
#Transpose your dataframe - Use .T

Origin_Cylinder.T
```

```python
#Plot transposed dataframe

sns.heatmap(Origin_Cylinder.T)
```

```python
# check correlations for cars dataset

cars.corr() # this gives you the Pearson correlation coefficient
```

```python
#Build heatmap of correlation matrix

sns.heatmap(cars.corr())
```

```python
#Seaborn default color palette maps lighter colors to larger 
#values and darker colors to smaller values even though
#the human eye tends to grasp darker colors first.

#We can change the color palette to match human behavior
#If you want to highlight those larger values with darker colors, you can do so


#Seaborn offers 170 different color palettes

```

```python
#Start with a SEQUENTIAL palette - reference cmap (color map)
# A Sequential palette means one color that varies from its lightest shade
#to its darkest shade

#Sequential palettes might be helpful if you want to highlight
#mostly large values or higher densities

sns.heatmap(Origin_Cylinder, cmap="Blues")

```

```python
#See heatmap for correlation matrix using a SEQUENTIAL palette

sns.heatmap(cars.corr(), cmap="Blues")
```

```python
sns.heatmap(cars.corr(), cmap="Greens")
```

```python
#Lets work with a DIVERGING palette
#Use a color palette that starts with Red and ends up with Blue

#Diverging color palettes might be helpful if you want to 
#highlight both large values and small values

sns.heatmap(cars.corr(), cmap='RdBu')
```

```python
sns.heatmap(cars.corr(), cmap = "PiYG")
```

```python
sns.heatmap(cars.corr(), cmap='coolwarm')
```

```python
#Ensuring center is zero when using a diverging color palette

sns.heatmap(cars.corr(), cmap='PiYG', center=0)
```

```python
#Setting the minimum and maximum values for 
#correlation heatmap
#Use the vmin and vmax arguments

sns.heatmap(cars.corr(), cmap='coolwarm',center=0, vmin=-1, vmax=1)
```

```python
#Annotate numbers directly on your heatmap in addition to colors
#Use annot argument

sns.heatmap(Origin_Cylinder, cmap='Greens', annot=True)
```

```python
#Get rid of scientific notation - and specify how many decimal points
#you would like Seaborn to display. Use fmt argument for this
#In this case we want a floating number with no decimal places

sns.heatmap(Origin_Cylinder,cmap='Blues', annot=True, fmt='.0f')
```

```python
# Specify that Seaborn use 1 decimal place instead of 0

sns.heatmap(Origin_Cylinder, cmap='Greens',annot=True, fmt='.1f')
```

```python
#Stlye your heatmap - font size

sns.heatmap(Origin_Cylinder,cmap='Blues', annot=True, fmt='.0f',
            annot_kws={'fontsize':16})
```

```python
#Style your heatmap - font size and font weight

sns.heatmap(Origin_Cylinder,cmap='Blues', annot=True, fmt='.0f',
            annot_kws={'fontsize':16, 'fontweight':'bold'})
```

```python
#Style your heatmap - font size, font weight and font family

sns.heatmap(Origin_Cylinder,cmap='Blues', annot=True, fmt='.0f',
            annot_kws={'fontsize':16, 'fontweight':'bold', 'fontfamily':'serif'})
```

```python
#Style your heatmap - add line between boxes
#Use linewidth argument

sns.heatmap(cars.corr(), cmap='coolwarm', linewidth=1)
```

```python
#Style your heatmap - add line between boxes
#Use linewidth argument

sns.heatmap(cars.corr(), cmap='coolwarm', linewidth=2)
```

```python
#Style your heatmap - add line between boxes and change line color
#Use linewidth argument
#Use linecolor argument

sns.heatmap(cars.corr(), cmap='coolwarm', linewidth=1, linecolor='black')
```

```python
#Get ticklabels to read better

#Create labels

labels = ['MPG','Cylinders','Displacement','Horsepower','Weight','Acceleration','Model Year']
```

```python
#Appply your label to your heatmap

sns.heatmap(cars.corr(), cmap='RdBu',
           xticklabels=labels,
           yticklabels=labels)
```

```python
#You can turn ticklabels for one axis off

sns.heatmap(cars.corr(), cmap='RdBu',
           xticklabels=False,
           yticklabels=labels)
```

```python
#You can have perfect squares for each square in your heatmap
#Use square argument

#Seaborn uses a square size based on the overall heatmap figure size as default
#To save this or any plot use the plt.savefig() function in the same cell as the code to generate the plot

sns.heatmap(cars.corr(), cmap='RdBu',
           xticklabels=labels,
           yticklabels=labels,
           square=True)

plt.savefig('C:/Users/abiy3759/Desktop/Colorado Boulder/INFO 5602- Information Visualization/test.png')
```
