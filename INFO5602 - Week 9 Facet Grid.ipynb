# Week 9, Day 1 - Seaborn Facet Grid

#WHAT IS IT?:


1. Seaborn's FacetGrid serves as the backbone for the following plots:

(1) catplot
(2) relplot
(3) displot


2. Seaborn uses the concept of small multiples to build out Facet Grids that in turn suppport various plots like scatterplots etc

#WHAT IS IT USED FOR?:

1. Facet Grid is used to create a grid, on which we can map other types of plots with minimal code


#WHEN IS IT USED?: 


1. When you need to build out various plots from scratch


```python
import seaborn as sns

from matplotlib import pyplot as plt
import numpy as np
```

```python
penguins = sns.load_dataset('penguins').dropna()
```

```python
penguins.head()
```

```python
sns.set_style('white')
```

```python
# Create initial Facet Grid with all dataset

sns.FacetGrid(penguins)
```

```python
# Supply column argument to build out the grid - using a particular categorical variable
#You can put charts on one row using the col argument

sns.FacetGrid(penguins, col='island')
```

```python
#Note that Facet Grid creates columns (or subplots) based on 
# the number of categories or values for that variable
# as found in the dataset
# For the variable island, Seaborn must have found 3 groups

```

```python
# We can also supply a row argument 
# to build out the grid - using a particular categorical variable

#You can put charts on one row using the rwo argument

sns.FacetGrid(penguins, row='island')
```

```python
# We can supply both column and row arguments to create a matrix 
# of rows and columns

sns.FacetGrid(penguins, col='island', row='species')
```

```python
# We can create a Facet Grid object - and use this to drive 
# the build out of the plots we want on the grid


# Lets create an object g to hold the Facet Grid

g=sns.FacetGrid(penguins, col='island')
```

```python
#Obtain the object type

type(g)

#We see that g is a FacetGrid object
```

```python
#We can employ methods of the FacetGrid object g to map/layer/insert plots on
# the empty Facet Grid we have built
# Use .map_dataframe()
#Specify the type of plot and the variable you want plotted

g=sns.FacetGrid(penguins, col='island')
g.map_dataframe(sns.histplot, 'flipper_length_mm')

#Note you should have both lines of code in the same cell

```

```python
# Change chart alignment - from columns to rows
#Use row instead of column argument

g=sns.FacetGrid(penguins, row='island')
g.map_dataframe(sns.histplot, 'flipper_length_mm')

#Note you should have both lines of code in the same cell
```

```python
#.map_dataframe allows for named variable arguments

g=sns.FacetGrid(penguins, row='island')
g.map_dataframe(sns.displot, 'flipper_length_mm')

#Note you should have both lines of code in the same cell
```

```python
#.map_dataframe allows for named variable arguments
#Use x = and y = to position the charts on the horizontal
# and vertical axis as you want

g=sns.FacetGrid(penguins, row='island')
g.map_dataframe(sns.histplot, x='flipper_length_mm')

#Note you should have both lines of code in the same cell
```

```python
#.map_dataframe allows for named variable arguments
#Use x = and y = to position the charts on the horizontal
# and vertical axis as you want

g=sns.FacetGrid(penguins, col='island')
g.map_dataframe(sns.histplot, y='flipper_length_mm')

#Note you should have both lines of code in the same cell
```

```python
#.map_dataframe allows for named variable arguments
#Use x = and y = to position the charts on the horizontal
# and vertical axis as you want

g=sns.FacetGrid(penguins, col='island')
g.map_dataframe(sns.histplot, x='flipper_length_mm')

#Note you should have both lines of code in the same cell
```

```python
#We can work with kdeplots

#Note that you use kdeplot and not kde

g=sns.FacetGrid(penguins, col='island')
g.map_dataframe(sns.kdeplot, x='bill_length_mm')

#Note you should have both lines of code in the same cell
```

```python
#We can also work with boxplots 

g=sns.FacetGrid(penguins, col='island')
g.map_dataframe(sns.boxplot, x='bill_length_mm')

#Note you should have both lines of code in the same cell
```

```python
#We can also work with scatterplots


g=sns.FacetGrid(penguins, col='island')
g.map_dataframe(sns.scatterplot, x='bill_length_mm', y='bill_depth_mm')

#Note you should have both lines of code in the same cell
```

```python
#There are three main steps to produce a Facet Grid with Seaborn:

#Step 1: Set up Facet Grid (sns.FacetGrid())

#Step 2: Specify plot type per facet (g.map_dataframe())

#Step 3: Customize your Facet Grid using methods and attributes
```

```python
#Methods and Attributes:

#The object g comes with methods that can be used to customize the Facet Grid

#Press Tab after g. to see the properties you can use

# g.
```

```python
#Lets use some methods to style the plot

#.set_axis_labels(): use to label the vertical and horizontal axes

g=sns.FacetGrid(penguins, col='island')
g.map_dataframe(sns.scatterplot, x='bill_length_mm', y='bill_depth_mm')
g.set_axis_labels('Bill Length (mm)', 'Bill Depth (mm)')
```

```python
#Lets use some methods to style the plot

#.set_axis_labels(): use to label the vertical and horizontal axes

# Here we supply a value for col and row arguments

g=sns.FacetGrid(penguins, col='island', row='species')
g.map_dataframe(sns.scatterplot, x='bill_length_mm', y='bill_depth_mm')
g.set_axis_labels('Bill Length (mm)', 'Bill Depth (mm)')
```

```python
#Using hue to breakout your chart with a categorical variable

#Remember to pass hue to the sns.FacetGrid line and not the
#.map_dataframe() line

#Also add legend with g.add_legend

g=sns.FacetGrid(penguins, col='island', hue='species')
g.map_dataframe(sns.scatterplot, x='bill_length_mm', y='bill_depth_mm')
g.set_axis_labels('Bill Length (mm)', 'Bill Depth (mm)')
g.add_legend()
```

```python
#Using hue to breakout your chart with a categorical variable

#Remember to pass hue to the sns.FacetGrid line and not the
#.map_dataframe() line

#Also add legend with g.add_legend
#Change color palette

g=sns.FacetGrid(penguins, col='island', hue='species', palette='prism')
g.map_dataframe(sns.scatterplot, x='bill_length_mm', y='bill_depth_mm')
g.set_axis_labels('Bill Length (mm)', 'Bill Depth (mm)')
g.add_legend()
```
