```python
#Dataset 1: For questions 1 - 4 below:

#Run the following before you attempt questions 1 - 4 below:

import seaborn as sns

from matplotlib import pyplot as plt

import numpy as np

import pandas as pd

from sklearn.datasets import load_breast_cancer
cancer_dataset= load_breast_cancer()

cancer_data = pd.DataFrame(np.c_[cancer_dataset['data'],cancer_dataset['target']],
             columns = np.append(cancer_dataset['feature_names'], ['target'])).dropna()

cancer_data.head()
```

```python
#Question 1

#Using Dataset 1 above write Seaborn code to build a blank pairplot using the variable "blank"
#as the left hand side of the code to generate the blank pairplot grid

#Use the following attributes/features for the plot:

#mean radius
#mean texture
#mean perimeter
#mean area


#Write your answer below:
blank = sns.PairGrid(cancer_data, vars=['mean radius', 'mean texture', 'mean perimeter', 'mean area'])

```

```python
#Question 2

#Write Seaborn code to insert histograms in the diagonals of the blank pairplot you developed above
#Set the transparency of your chart to 0.3

#Write your answer below:


blank.map_diag(plt.hist, alpha=0.3)
```

```python
#Question 3

#Write Seaborn code to insert scatterplots in the off-diagonals of the blank pairplot you developed above
#Set the transparency of your chart to 0.7

#Write your answer below:


blank.map_offdiag(plt.scatter, alpha=0.7)

```
